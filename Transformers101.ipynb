# Transformers-101

Untill now we made good progress about the models that can approximate *continuous* distribution for images for example. But what about *Cathegorical Distributions* ?

A Cathegorical distribution governs how variability occurs over a *finite* set of realizations, the probability space is finite in this case (not like $\mathcal{R}$ where images live), building such model for this case is a big deal inded, since many scenarios from our world are goverened by this type of distribution, the easiest example out there is **Text**.

<img src="docs/LLM/probs.png" width="700">


The way to see it is the following : imagine each word in our vocabulary a unique item, hence the overall set of possible items is the whoole vocabulary. and we want to build a model that understands the interaction of each word with all the other ones. such model will naturally understand without intervention how verbs occurs usualy after nouns, and each sentence is finished with a period ".", and so on ...

why weren't they present waay before, simply because text is very-difficult and complex distribution to capture for many reasons : 
- text is governed by many rules, that are hard to capture, and most words highly depends on their context, take the example of "right" : where it can mean the opposite of "left", the fact of being "correct", or someone's "rights" in juredic, (same for so many words : *run, charge, match, bat* ....)
- second, the number of possible words is just **huge**, with all the engineering that can be done (sub-tokeniation..) let's say 171,000 words in current use according to oxford Dictionary, imagining the easier scenario where our vocabulary, (the set of words $\Omega = {cat, dog, house....} $the number of words is $Card(\sigma)=10^{10}$, we make the n-gram assumtion : the occuring porbability of $i_th$ word in a sentence depends only on the n-word behind it, mathematically : 
$\mathcal{P}(word_i|sentence) = \mathcal{P}(word_i|word_{i-1}, word_{i-2}, ..., word_{i-n}) $.
Putting it all together, for a naïve case where n = 5 (extremely upracticall, where in real scenarios words in a text depend on what is said many  sentences before sometimes), our desired model must find the Probability-value of eqaution (1) of all the possible words, which makes $10^{10} * 10^{10} * ...= 10^{50}$ parameters to find and store somewhere, that's approxiamtely the number of atoms in the universe.

That's why language modeling was a very complex and scary task to perform by AI, untill the appearence of the transformer's architecture, that soved the problem by introducing a smart modelisation and training solutions to outsmart the text modeling.
